# Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.distplot([1,2,3,4,5,6],hist=False) # distplot - distribution plot | without using histogram
plt.show()
```

## Normal distribution

```python
# loc - mean / scale - standard deviation
sns.distplot(random.normal(loc=1,scale=1,size=(2,3)),hist=False)
plt.show()
```

## Binomial distribution

```python
# n - no of trails, p- prob of occurance in each trials
sns.distplot(random.binomial(n=10,p=0.5, size=100))
plt.show()
```

## Similarities between Normal and Binomial Distribution

```python
sns.distplot(random.normal(loc=50,scale=5,size=1000),hist=False,label="normal") # normal
sns.distplot(random.binomial(n=100,p=0.5,size=1000),hist=False,label="binomial") #binomial
plt.xlabel('Quite similar in some cases')
plt.show()
```

## Poisson distribution

```python
sns.distplot(random.poisson(lam=4,size=1000),kde=False) # kernal density estimator
plt.show()  
```

## Uniform distribution

```python
# a- lower bound/ b - upper bound
sns.distplot(random.uniform(size=1000),hist=False)
plt.show()
```

## Logistic Distribution

```python
sns.distplot(random.logistic(size=1000),hist=False,label='logistic') # Blue - has more spread in tail
sns.distplot(random.normal(size=1000),hist=False,label='normal') # orange - has less spread in tail than logistic
plt.show()
```

## Multinomial distribution

```python
mul_nomial = random.multinomial(n=4,pvals=[1/4,1/6,1/6,1/6,1/6,1/6])
print(mul_nomial)
sns.distplot(mul_nomial,hist=False)
plt.show()
```

## Exponential distribution

```python
sns.distplot(random.exponential(scale=2,size=1000),hist=False)
plt.show()
```

## ChiSqaure distribution - important

```python
sns.distplot(random.chisquare(df=2,size=1000),hist=False)
plt.show()
```

## rayleigh distribution

```python
sns.distplot(random.rayleigh(scale=2,size=(1000)),hist=False)
plt.show()
```

## Pareto distribution

```python
sns.distplot(random.pareto(a=2,size=1000))
plt.show()
```

## Zipf distribution

```python
x = random.zipf(a=2,size=1000)
sns.displot(x[x<10],kde=False)
plt.show()
```
