```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Basic Graph doc - https://matplotlib.org/stable/plot_types/index

```python
x = [0,1,2,3,4,5]
y = [0,2,4,6,8,10]
     
    
# Resize the Graph
plt.figure(figsize=(5,3) , dpi=140)

# Forecasting
x2 = np.arange(0,4.5,0.5)
print(len(x2))
plt.plot(x2[:6], x2[:6]**2, label='Sqaured_values' , color='grey') # Normal
# Forecasting
plt.plot(x2[5:],x2[5:]**2, color='blue', linestyle='--')

plt.plot(x,y, label='2x', color="Orange", linewidth=2, linestyle='--', marker='*', markersize=12, markeredgecolor='blue')
plt.title("Linear Line", fontdict={'fontname':'Comic Sans MS','fontsize':20})

#labelling
plt.xlabel("X-axis")
plt.ylabel("Y-axis") 

# Naming
plt.xticks([0,1,2,3,4,5]) # For x axis - line value
# plt.yticks([4,5,6]) # For y axis - line value 

## Styling
plt.legend()
plt.savefig('my+graph.png', dpi=300)
plt.show()
```

## Bar Chart

```python
labels = ['A','B','C','D']
values = [1,2,3,2]

bar = plt.bar(labels, values)

pattern = ['/','o','*','/']
for bar in bar:
    bar.set_hatch(pattern.pop(0))
plt.show()
```
