```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
planets = sns.load_dataset('planets')
```

```python
tips.head()
```

## Scatterplot

```python
sns.scatterplot(x='tip', y='total_bill', data = tips, hue='day', size='size')
```

## Histogram 

```python
sns.histplot(tips['tip'], kde=True, bins=15)
```

## Bar plot

```python
 sns.barplot(x='sex',y='tip', data=tips)
```

## Box plot

```python
sns.boxplot(x='day',y='tip',data = tips, hue='sex')
```

## Strip plot

```python
sns.stripplot(x='day',y='tip', data=tips, hue='sex' , dodge=True) # Dodge = True - Means seperate hue categorical values withou colapsing
```

## Joint plot

```python
sns.jointplot(x='tip',y='total_bill', data=tips, kind='reg')
```

```python
# Kind kde
sns.jointplot(x='tip',y='total_bill', data=tips, kind='kde', fill=True)
```

```python
# Kind hex
sns.jointplot(x='tip',y='total_bill', data=tips, kind='hex')
```

## Pair plot

```python
sns.pairplot(titanic.select_dtypes(['number']),hue='pclass')
```

## Heatmap

```python
titanic.corr(numeric_only=True)
```

```python
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='coolwarm')
```

## Cluster map

```python
sns.clustermap(iris.drop('species',axis=1))
```
