```python
import numpy as nup
import pandas as pd
import matplotlib.pyplot as plt
import warnings
```

## Line graph

```python
df1 = pd.read_csv('data_set/gas_prices.csv')
df1.columns
```

```python
plt.figure(figsize=(8,6))
plt.title('Gas Price Prediction', fontdict={'fontname':'Comic Sans MS', 'fontsize':'20'})
plt.plot(df1['Year'],df1['Australia'], color='orange', label='For Australia')
plt.plot(df1['Year'],df1['Canada'], color='Blue', label='For Canada')
plt.xticks(df1['Year'][::3], rotation='vertical')
plt.legend()
plt.xlabel('Years')
plt.ylabel('Prices')

plt.savefig('data_set/Gas_Price_Prediction.png',dpi=300)

plt.show()

```

```python
## Printing all the country
plt.figure(figsize=(7,5))
for country in df1:
    if country != "Year":
        plt.plot(df1['Year'], df1[country] , marker='.')
plt.legend
plt.show()
```

## Load Fifa Data

```python
df = pd.read_csv('data_set/fifa_data.csv')
df.columns
```

# Histogram

```python
bins = list(range(10,101,10)) # 10 20 ... 100 interval
plt.hist(df['Overall'], bins = bins , color='r')
plt.xticks(bins)
plt.xlabel('Overall Skill Rating')
plt.ylabel('No of Players')
plt.title('Distribution of Fifa Players Skills')
plt.show()
```

 # Pie Chart

```python
left = df.loc[df['Preferred Foot'] == 'Left'].count()[0]
right = df.loc[df['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left','Right']
plt.pie([left,right], labels=labels, autopct='%.2f %%')
plt.title("Foot preferences of FIFA Players")
plt.show()
```

## Pie Chart 2

```python
df.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in df.Weight]
light = df.loc[df['Weight'] < 125].count()[0]
light_medium = df.loc[(df['Weight'] >= 125) & (df['Weight'] < 150)].count()[0]
medium = df.loc[(df['Weight'] >= 150) & (df['Weight'] < 175)].count()[0]
medium_heavy = df.loc[(df['Weight'] >= 175) & (df['Weight'] < 200)].count()[0]
heavy = df.loc[(df['Weight'] >= 200)].count()[0]

labels = ['<125','125-150','150-175','175-200','>200']

explode = (.2,.1,0,0,.2)

plt.pie([light,light_medium,medium,medium_heavy,heavy], labels=labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)

plt.title("Weight Distribution of Fifa Players")

plt.show()
```

# Bar Plot

```python
barcelona = df.loc[df['Club'] == 'FC Barcelona']['Overall']
real_madrid = df.loc[df['Club'] == 'Real Madrid']['Overall']
new_eng = df.loc[df['Club'] == 'New England Revolution']['Overall']

plt.figure(figsize=(5,6))

box = plt.boxplot([barcelona, real_madrid, new_eng] ,labels=['Barcelona','Real Madrid', 'New England Revolution'])

plt.title('Fifa Team Comparision for Allotment')
plt.xlabel('Clubs')
plt.ylabel('Overall')
plt.show()
```
